## Task 1: Load the Data

```python
# We will start by importing the necessary libraries
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
# Directly download the Excel file instead of using a redirect link
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
df = pd.read_excel(url, engine='openpyxl')  # Explicitly specify the engine
```

## Task 2: Perform data cleaning

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Remove rows with any missing values
df.dropna(inplace=True)

# Remove columns with any missing values
df.dropna(axis=1, inplace=True)
```

```python
# Fill missing values in 'Quantity' with the mean
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
```

```python
# Forward fill missing values in 'UnitPrice'
df['UnitPrice'].fillna(method='ffill', inplace=True)
```

```python
# Drop the 'CustomerID' and 'StockCode' columns
df.drop(['CustomerID', 'StockCode'], axis=1, inplace=True)
```

```python
# Impute missing values in 'Quantity' with the mean
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)  # Replace NaN with mean

# Impute missing values in 'UnitPrice' with the median
df['UnitPrice'].fillna(df['UnitPrice'].median(), inplace=True)  # Replace NaN with median
```

# Task 3: Explore the basic statistics of the dataset, including measures of central tendency and dispersion.

```python
# Calculate mean, median, and mode for 'UnitPrice'
mean_price = df['UnitPrice'].mean()
median_price = df['UnitPrice'].median()
mode_price = df['UnitPrice'].mode()[0]  # mode() returns a Series, get the first value

print(f"Mean UnitPrice: {mean_price}")
print(f"Median UnitPrice: {median_price}")
print(f"Mode UnitPrice: {mode_price}")
```

```python
# Calculate range, variance, and standard deviation for 'Quantity'
range_quantity = df['Quantity'].max() - df['Quantity'].min()
variance_quantity = df['Quantity'].var()
std_dev_quantity = df['Quantity'].std()

print(f"Range of Quantity: {range_quantity}")
print(f"Variance of Quantity: {variance_quantity}")
print(f"Standard Deviation of Quantity: {std_dev_quantity}")
```

# Task 5: Analyze the sales trends over time. Identify the busiest months and days of the week in terms of sales.

```python
# Extract month from InvoiceDate
df['InvoiceMonth'] = df['InvoiceDate'].dt.month

# Group by month and sum total sales
monthly_sales = df.groupby('InvoiceMonth')['Total_Sales'].sum().reset_index()

# Create the interactive bar chart
fig_monthly_sales = px.bar(monthly_sales,
                           x='InvoiceMonth',
                           y='Total_Sales',
                           title='Monthly Sales Trend',
                           labels={'InvoiceMonth': 'Month', 'Total_Sales': 'Total Sales'})

# Customize x-axis labels to show month names
fig_monthly_sales.update_xaxes(tickvals=list(range(1, 13)),
                               ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                         'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

fig_monthly_sales.show()
```

**We see November is the busiest month**

```python
# Calculate daily sales
daily_sales = df.groupby(df['InvoiceDate'].dt.dayofweek)['Total_Sales'].sum()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_sales.index, y=daily_sales.values)

# Customize the plot
plt.title('Sales Trend by Day of the Week')
plt.xlabel('Day of Week (0=Mon, 6=Sun)')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

plt.show()
```

**We see that Thursday is the busiest day**.

```python
# 3. Identify Busiest Month and Day
# Another code
# Find the month with the highest sales
busiest_month = monthly_sales.loc[monthly_sales['Total_Sales'].idxmax(), 'InvoiceMonth']

# Find the day of the week with the highest sales
busiest_day = daily_sales.loc[daily_sales['Total_Sales'].idxmax(), 'InvoiceDayOfWeek']

print(f"Busiest Month: {busiest_month} (corresponds to month name using the mapping above)")
print(f"Busiest Day of the Week: {busiest_day} (0 = Monday, 6 = Sunday)")
```

To display the 10 most popular procduct for each country by using a Drop-Down List


```python
top_n = 10

# Group data by product description and country, and sum quantities
product_sales_by_country = df.groupby(['Country', 'Description'])['Quantity'].sum().reset_index()
product_sales_by_country.rename(columns={'Quantity': 'TotalQuantity'}, inplace=True)

# Get unique countries
countries = product_sales_by_country['Country'].unique()

# Create initial figure (start with the first country)
initial_country = countries[0]
initial_data = product_sales_by_country[product_sales_by_country['Country'] == initial_country].nlargest(top_n, 'TotalQuantity')

fig = px.bar(initial_data,
             x='Description',
             y='TotalQuantity',
             title=f"Top {top_n} Most Popular Products in {initial_country}",
             labels={'TotalQuantity': 'Total Quantity Sold'})
fig.update_xaxes(tickangle=45, tickfont=dict(size=10))

# Add dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=[{'x': [product_sales_by_country[product_sales_by_country['Country'] == country].nlargest(top_n, 'TotalQuantity')['Description']],
                           'y': [product_sales_by_country[product_sales_by_country['Country'] == country].nlargest(top_n, 'TotalQuantity')['TotalQuantity']]},
                          {'title': f"Top {top_n} Most Popular Products in {country}"}],
                    label=country,
                    method="update"
                ) for country in countries
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)

fig.show()
```

# Task 7: Identify any outliers or anomalies in the dataset and discuss their potential impact on the analysis.

1. Identifying Outliers in the Online Retail Dataset

---


We're specifically interested in identifying outliers related to customer purchases in this project. These outliers might manifest in variables like "UnitPrice" (products with unusually high or low prices) and "Quantity" (orders with unusually large quantities).

Here's how we can identify these outliers:

Visualizing Outliers with Box Plots: Creating box plots for "UnitPrice" and "Quantity" will give us a quick visual representation of potential outliers. Outliers will be plotted as individual points beyond the "whiskers" of the box plot, indicating values significantly different from the typical range of the data.
Statistical Approaches for Outlier Detection:
Z-scores: Calculating Z-scores for "UnitPrice" and "Quantity" can help identify data points that are far from the average in terms of standard deviations. For example, a product with a Z-score of 3 for "UnitPrice" would have a price three standard deviations above the average, making it a potential outlier.
Interquartile Range (IQR): Using the IQR method, we can set boundaries for typical data values. Any value falling outside these boundaries (below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR) would be considered an outlier. This method is particularly useful for skewed distributions.
2. Impact of Outliers on Online Retail Analysis

---


Outliers in this dataset can have significant implications for understanding the business:

Skewed Sales Metrics: Outliers in "UnitPrice" or "Quantity" can significantly impact calculations of total sales, average order value, and product popularity. For example, a single very large order could inflate the average order value, giving a misleading impression of typical customer behavior.
Misinterpretation of Customer Behavior: Outliers might represent unusual purchasing patterns, such as bulk orders from businesses or errors in data entry. Failing to account for these outliers could lead to incorrect assumptions about customer preferences and buying habits.
Inaccurate Business Decisions: If outliers significantly influence your analysis, they could lead to flawed decisions regarding inventory management, pricing strategies, and marketing campaigns. For instance, stocking up on a product with a few outlier purchases might not be a profitable strategy.
3. Handling Outliers in the Online Retail Context

---


Given the potential impact of outliers on our analysis, we need to carefully consider how to address them:

Investigating Outlier Causes: Before making any changes to the data, it's crucial to understand why the outliers exist. Are they data entry errors, legitimate large orders, or something else? This investigation will inform the best approach for handling them.
Removal with Justification: If outliers are clearly due to errors or represent events outside the scope of our analysis (e.g., wholesale orders when focusing on retail customers), it might be appropriate to remove them. However, this decision should always be documented and justified.
Data Transformation: Applying a log transformation to "UnitPrice" or "Quantity" can help reduce the influence of extreme values and bring the data closer to a normal distribution. This can be helpful for certain statistical analyses.
Segmentation or Separate Analysis: If outliers represent meaningful segments of the data (e.g., bulk buyers), it might be better to analyze them separately or segment the data to gain insights into different customer groups.
By thoughtfully addressing outliers in the Online Retail dataset, we can ensure that our analysis provides accurate and actionable insights for the business. Remember to carefully document your approach and consider the implications of your decisions on the final conclusions.

```python
my_dict = {'allocation %': [20, 20, 20, 20, 20]   } #remove one element to match the length of the index
crypto_df = pd.DataFrame(data = my_dict, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
crypto_df
```

```python
fig = px.pie(crypto_df, values = 'allocation %', names = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'], title = 'Crypto Allocation')
fig.show()
```

```python
my_dict = {'allocation %': [10 ,10, 10, 60, 10]  }
crypto_df = pd.DataFrame(data = my_dict, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
fig = px.pie(crypto_df, values = 'allocation %', names = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'], title = 'Crypto Allocation')
fig.show()
```

```python
fig = px.pie(crypto_df, values = 'allocation %', names = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'], hole=0.3, title = 'Crypto Allocation')
fig.show()
```

```python
job_1 = {'Task':'Develop content', 'Start':'2020-10-10', 'Finish':'2021-01-01'}
job_1
```

```python
job_2 = {'Task':'Film Videos', 'Start':'2021-01-01', 'Finish':'2021-03-03'}
job_2
```

```python
job_3 = {'Task':'Edit Videos', 'Start':'2021-03-03', 'Finish':'2021-04-04'}
job_3
```

```python
project = pd.DataFrame([job_1, job_2, job_3])
project
```

```python
fig = px.timeline(project, x_start = 'Start', x_end = 'Finish', y = 'Task')
fig.update_yaxes(autorange = 'reversed')
fig.show()
```

```python
job_4 = {'Task':'Send the Course for Approval', 'Start':'2021-04-04', 'Finish':'2021-04-05'}
job_4

project = pd.DataFrame([job_1, job_2, job_4])
project

fig = px.timeline(project, x_start = 'Start', x_end = 'Finish', y = 'Task')
fig.update_yaxes(autorange = 'reversed')
fig.show()
```
